```python
def certigen(Name):
    img = cv2.imread(r"D:\IDC\certificate generator\valorent.jpg")
    value =int(len(Name)/2)+1
    xval= (1240-(value*46)) 
    font =cv2.FONT_HERSHEY_SIMPLEX
    org = (xval, 1800)
    fontScale = 3
    color = (255,255,255)
    thickness = 7
    image = cv2.putText(img,Name, org, font,fontScale, color, thickness)
    filename =Name+".jpg" 
    cv2.imwrite(filename,img)
```
